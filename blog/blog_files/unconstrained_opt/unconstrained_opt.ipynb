```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

<h1>Graph the function from the blog post</h1>

```python
# Define the function
f = lambda x, y: -4*x**2 - 2*y**2

# Define the linspace to plot.
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)

# Create the meshgrid
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Define our plotting stuff
ax = plt.figure(dpi=150).add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8, alpha=0.3)

# Plot the projections of the contours for each dimension.
ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap='coolwarm')
ax.contourf(X, Y, Z, zdir='x', offset=-3.5, cmap='coolwarm')
ax.contourf(X, Y, Z, zdir='y', offset=3.5, cmap='coolwarm')

# Set the axes limits
ax.set(xlim=(-3.5, 3.5), ylim=(-3.5, 3.5), zlim=(-100, 0), xlabel=r'X', ylabel=r'Y', zlabel=r'Z', title=r'$f(x, y) = -4x^2 - 2y^2$')
plt.show()
```

<h1>The Rosenbrock Function!</h1>

```python
f = lambda x, y: (1 - x)**2 + 100*(y - x**2)**2

# Define the linspace to plot.
y = np.linspace(-.7, 2.8, 300)
x = np.linspace(-2, 2, 300)

# Create the meshgrid
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Define our plotting stuff
ax = plt.figure(dpi=150).add_subplot(projection='3d')
ax.plot_surface(Y, X, Z, cmap='viridis')

# Set the axes limits
ax.set(xlim=(-1, 3), ylim=(-2, 2), zlim=(0, 2000), xlabel=r'Y', ylabel=r'X', zlabel=r'Z', title=r'$f(x, y) = (1 - x)^2 + 100(y - x^2)^2$')
ax.set_xticks(np.arange(-1, 4, 1), label=[-1, 0, 1, 2, 3])
ax.set_yticks(np.arange(-2, 3, 1), label=[-2, -1, 0, 1, 2])
ax.set_zticks(np.linspace(0, 2000, 5), label=[0, 500, 1000, 1500, 2000])
ax.set_zbound(0, 2000)
ax.set_ylim(ax.get_ylim()[::-1])
plt.show()
```

```python
f = lambda x, y: (1 - x)**2 + 100*(y - x**2)**2

# Define the linspace to plot.
y = np.linspace(-.7, 2.8, 300)
x = np.linspace(-2, 2, 300)

# Create the meshgrid
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Define our plotting stuff
fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(Y, X, Z, cmap='viridis')

# Set the axes limits
ax.set(xlim=(-1, 3), ylim=(-2, 2), zlim=(0, 2000), xlabel=r'Y', ylabel=r'X', zlabel=r'Z')
ax.set_xticks(np.arange(-1, 4, 1), label=[-1, 0, 1, 2, 3])
ax.set_yticks(np.arange(-2, 3, 1), label=[-2, -1, 0, 1, 2])
ax.set_zticks(np.linspace(0, 2000, 5), label=[0, 500, 1000, 1500, 2000])
ax.set_zbound(0, 2000)
ax.set_ylim(ax.get_ylim()[::-1])
ax.view_init(elev=30, azim=-40)
ax.plot([1], [1], f(1,1), marker='o', markersize=5, color='red', zorder=10)


ax = fig.add_subplot(1, 2, 2)
# Define the linspace to plot.
y = np.linspace(-1, 4, 300)
x = np.linspace(-2, 2, 300)

# Create the meshgrid
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.contourf(X, Y, Z, cmap='viridis')

# Set the axes limits
ax.set(xlim=(-2, 2), ylim=(-1, 4), xlabel=r'X', ylabel=r'Y')
ax.set_xticks(np.arange(-2, 2, 1), label=[-2, -1, 0, 1, 2])
ax.set_yticks(np.arange(-1, 4, 1), label=[-1, 0, 1, 2, 3, 4])
# ax.set_ylim(ax.get_ylim()[::-1])
ax.scatter(1, 1, marker='o', s=20, color='red', zorder=10, label=r'(1,1)')
legend = ax.legend(title='Minimum', loc='upper right')

plt.suptitle('The Rosenbrock Function')
plt.tight_layout(w_pad=5)
plt.show()
```
